```python
dmI=dm.iloc[:,1:]
dmI.head()
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
wcss=[]
for i in range (1,12):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(dmI)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,12),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
print(wcss, len(wcss))
```

```python
dm.shape
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=42)
```

```python
model.fit(dmI)
```

```python
label=model.labels_
print(label, len(label))
```

```python
dm0=pd.read_csv("main_data.csv") # original with country codes
dm['country']=dm0['country']
dm.head()
```

```python
label
```

```python
least=[]
moderate=[]
highest=[]

for i in range(len(dm)):
    if label[i]==1:
        highest.append(dm['country'][i])
    elif label[i]==2:
        least.append(dm['country'][i])
    else:
        moderate.append(dm['country'][i])
```

```python
print("Countries with least regulatory performance: \n",least)
print(len(least))
```

```python
print("Countries with reasonable regulatory performance: \n",moderate)
print(len(moderate))
```

```python
print("Countries with highest regulatory performance: \n",highest)
print(len(highest))
```
